```python
df = pd.read_excel('GLTab27072021.xlsx', engine='openpyxl')
#df['As Quarter'] = df['Posting Date'].dt.to_period('Q-MAR')
#df['Fiscal Year'] = df['As Quarter'].dt.qyear
df['Fiscal Year'] = df['Posting Date'].dt.to_period('Q-APR').dt.qyear.apply(lambda x: str(x-1) + "-" + str(x))
df
```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
FiscalYear = list(df['Fiscal Year'].unique())
#Year = widgets.Dropdown(options=['hello', 'world'], value=None)
Year = widgets.Dropdown(options=FiscalYear, value=None)
Submit = widgets.Button(description="Update")
#def update():
#    #display(HBox([Year, Submit]))
#    print("1")
#    #display(py.offline.iplot(data, filename='jupyter-basic_bar'))

def on_button_clicked(b):
    condition = (df['G/L Account'] > 5000000) & (df['G/L Account'] < 5799999) & (df['Business Area'] != 4048)
#    condition = condition & (df['Fiscal Year'] == '2020-2021')
    condition = condition & (df['Fiscal Year'] == Year.value)
    PivotTable = df.loc[condition].pivot_table(index=['Fiscal Year', 'Business Area'],
                columns='G/L Account', values='Amount in local currency', aggfunc='sum', margins = True).div(10000000)
    PivotTable = PivotTable.round(2)
    clear_output()
    display(PivotTable)
    display(Year, Submit)

display(Year, Submit)
Submit.on_click(on_button_clicked)
```
